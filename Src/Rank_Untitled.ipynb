```python
import pandas as pd
from pymongo import MongoClient
import fn_mongo as fm
```

```python
comps_coll = fm.connect_collection('companies', 'companies')
```

```python
companies = comps_coll.find()
```

```python
df_comps = pd.DataFrame(companies)
df_comps.head()
```

```python
lst_loc = [df_comps.location[e] for e in range(len(df_comps))]
lst_loc[0:2]
```

```python
radio = 5000
query = {'raised_dollar':{'$gte':1000000}}
lst_num_offices = []
lst_offices = []
for i in range(len(df_comps)):
    num_off = fm.find_offices_near(comps_coll, lst_loc[i], query, radio)
    lst_offices.append(num_off)
    lst_num_offices.append(len(num_off))
    
df_comps['number offices'] = lst_num_offices
```
